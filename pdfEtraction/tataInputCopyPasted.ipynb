```python
import pandas as pd

input_df = pd.read_csv(r"D:\Nexensus_Projects\pdfEtraction\tataInput.csv")

input_df['Pan'] = "AAATT0570A"
input_df['ISIN No.'] = None
input_df['Mutual Fund'] = "TATA MUTUAL FUND"
input_df['Fund Type'] = input_df['Fund Name'].apply(lambda x: x.split(' ', 1)[-1])
input_df['Portfolio'] = input_df['Name of the Instrument']
input_df['Agency'] = input_df['Agency_and_Ratings'].apply(lambda x: None if len(str(x).split(" "))==1 else str(x).split(" ")[0])
input_df['Ratings'] = input_df['Agency_and_Ratings'].apply(lambda x: str(x).split(" ")[-1])
input_df['Category'] = None
input_df['Factsheet Date'] = None
input_df["URL"] = None
```

```python
import re
# Save correct_rating_df to Excel file with separate sheets for each 'fund'
output_df = input_df[["Pan", "ISIN No.", "Mutual Fund", "Fund Name", "Fund Type", "Portfolio", "Agency",
                       "Ratings", "Category",	"(%) Of Total AUM", "Factsheet Date", "URL",]]

output_excel = "tataOutput.xlsx"

with pd.ExcelWriter(output_excel, engine='openpyxl') as writer:
    for fund, group_df in output_df.groupby("Fund Name"):
        # Clean sheet name: max 31 characters, remove invalid characters
        sheet_name = str(fund)[:31]
        invalid_chars = r'[:\\/?*[\]]'
        sheet_name = re.sub(invalid_chars, "_", sheet_name)

        print(f"Writing sheet: {sheet_name}")
        group_df.to_excel(writer, sheet_name=sheet_name, index=False)
        # group_df.drop(columns=["Fund Name"], errors="ignore").to_excel(writer, sheet_name=sheet_name, index=False)
        print("--------------------------------------------")

print(f"✅ Processed data saved to: {output_excel}")
```
