```python
############## Code to split pdf to keep only pages with Rating table in it #####################
import pdfplumber
from PyPDF2 import PdfWriter, PdfReader
from datetime import datetime

# month_name = datetime.now().strftime("%b")
month_name = "July"

input_pdf_path = "BBNPP_MF_Fund_Facts_June_2025_13196.pdf"
output_pdf_path = "filtered_{}.pdf".format(input_pdf_path.split(".")[0])

reader = PdfReader(input_pdf_path)
writer = PdfWriter()
fund_list = []

with pdfplumber.open(input_pdf_path) as pdf:
    for page_num, page in enumerate(pdf.pages):
        tables = page.extract_tables()

        for table in tables:
            # header = table[0]
            for t in table:
                if "Rating" in t or 'ating' in t:
                    print(table)
                    writer.add_page(reader.pages[page_num])
                    continue  # add this page once, then move to next page
        print(page_num+1, ">>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")



# Save the filtered PDF
with open(output_pdf_path, "wb") as f_out:
    writer.write(f_out)

print(f"✅ New PDF created: {output_pdf_path}")
```
