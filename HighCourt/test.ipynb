```python
################### Code to separate company payloads and non company payloads ###########################################

import os
import json
import pandas as pd

# Base input and output root directories
base_input_dir = r'D:\Nexensus_Projects\HighCourt\High_court_data\Hightcourt_payload'
base_entity_output_dir = r'D:\Nexensus_Projects\HighCourt\High_court_data\Highcourt_company_payload'
base_non_entity_output_dir = r'D:\Nexensus_Projects\HighCourt\High_court_data\Highcourt_noncompany_payload'

# Entity matching keywords
keywords = ['llp', 'ltd', 'bank', 'limited']

def contains_entity_keyword(name):
    if pd.isna(name):
        return False
    name = name.lower()
    return any(keyword in name for keyword in keywords)

# Loop through all years and categories dynamically
for year in os.listdir(base_input_dir):
    year_path = os.path.join(base_input_dir, year)
    if not os.path.isdir(year_path):
        continue

    for category in os.listdir(year_path):
        category_path = os.path.join(year_path, category)
        if not os.path.isdir(category_path):
            continue

        print(f"\n📁 Processing: Year = {year}, Category = {category}")

        # Define dynamic output directories
        entity_output_dir = os.path.join(base_entity_output_dir, year, category)
        non_entity_output_dir = os.path.join(base_non_entity_output_dir, year, category)
        os.makedirs(entity_output_dir, exist_ok=True)
        os.makedirs(non_entity_output_dir, exist_ok=True)

        for filename in os.listdir(category_path):
            if filename.endswith('.txt'):
                input_path = os.path.join(category_path, filename)
                with open(input_path, 'r', encoding='utf-8') as f:
                    lines = f.readlines()

                data = []
                for line in lines:
                    line = line.strip().rstrip(',')
                    if line:
                        try:
                            obj = json.loads(line)
                            data.append(obj)
                        except json.JSONDecodeError as e:
                            print(f"⚠️ Skipping invalid JSON in {filename}: {e}")

                if not data:
                    continue

                df = pd.DataFrame(data)
                df['isPetitionerEntity'] = df['petitionerName'].apply(contains_entity_keyword)
                df['isRespondentEntity'] = df['respondentName'].apply(contains_entity_keyword)

                non_entities = df[(df['isPetitionerEntity'] == False) & (df['isRespondentEntity'] == False)]
                entities = df[~((df['isPetitionerEntity'] == False) & (df['isRespondentEntity'] == False))]

                # Output file paths
                entity_path = os.path.join(entity_output_dir, filename)
                non_entity_path = os.path.join(non_entity_output_dir, filename)

                with open(non_entity_path, 'w', encoding='utf-8') as f1:
                    for record in non_entities.drop(columns=['isPetitionerEntity', 'isRespondentEntity']).to_dict(orient='records'):
                        f1.write(json.dumps(record) + '\n')

                with open(entity_path, 'w', encoding='utf-8') as f2:
                    for record in entities.drop(columns=['isPetitionerEntity', 'isRespondentEntity']).to_dict(orient='records'):
                        f2.write(json.dumps(record) + '\n')

                print(f"✅ Processed: {filename}")

print("\n🎉 All dynamic folders processed successfully.")
```

```python
########################### Code for checking count of noncompany and company payload from original number of payloads ####
import os
from collections import defaultdict

def get_file_line_counts_by_year_and_category(base_dir):
    counts = defaultdict(lambda: defaultdict(dict))  # {year: {category: {filename: count}}}
    for root, _, files in os.walk(base_dir):
        for file in files:
            if file.endswith('.txt'):
                rel_path = os.path.relpath(root, base_dir)
                parts = rel_path.split(os.sep)
                if len(parts) >= 2:
                    year, category = parts[0], parts[1]
                    full_path = os.path.join(root, file)
                    try:
                        with open(full_path, 'r', encoding='utf-8') as f:
                            count = sum(1 for _ in f if _.strip())
                        counts[year][category][file] = count
                    except Exception as e:
                        print(f"⚠️ Failed to read {full_path}: {e}")
    return counts

# Define all 3 directories
input_dir = r'D:\Nexensus_Projects\HighCourt\High_court_data\Hightcourt_payload'
entity_dir = r'D:\Nexensus_Projects\HighCourt\High_court_data\Highcourt_company_payload'
non_entity_dir = r'D:\Nexensus_Projects\HighCourt\High_court_data\Highcourt_noncompany_payload'

# Get counts from all sources
input_counts = get_file_line_counts_by_year_and_category(input_dir)
entity_counts = get_file_line_counts_by_year_and_category(entity_dir)
non_entity_counts = get_file_line_counts_by_year_and_category(non_entity_dir)

# Display summary
for year in sorted(set(input_counts.keys()) | set(entity_counts.keys()) | set(non_entity_counts.keys())):
    print(f"\n📅 Year: {year}")
    categories = set(input_counts.get(year, {}).keys()) | set(entity_counts.get(year, {}).keys()) | set(non_entity_counts.get(year, {}).keys())

    for category in sorted(categories):
        print(f"\n🔹 Category: {category}")
        print(f"{'File Name':<40} {'Input':>8} {'Entity':>10} {'Non-Entity':>12} {'Total Out':>12} {'Diff':>8}")
        print("-" * 95)

        files = set(input_counts.get(year, {}).get(category, {}).keys()) | \
                set(entity_counts.get(year, {}).get(category, {}).keys()) | \
                set(non_entity_counts.get(year, {}).get(category, {}).keys())

        for file in sorted(files):
            input_count = input_counts.get(year, {}).get(category, {}).get(file, 0)
            entity_count = entity_counts.get(year, {}).get(category, {}).get(file, 0)
            non_entity_count = non_entity_counts.get(year, {}).get(category, {}).get(file, 0)
            total_out = entity_count + non_entity_count
            diff = input_count - total_out
            print(f"{file:<40} {input_count:>8} {entity_count:>10} {non_entity_count:>12} {total_out:>12} {diff:>8}")
```

```python
import re

def normalize_company_name(company_name: str) -> str:
    name = company_name.strip().title()

    # Replace common abbreviations and misformatted versions
    substitutions = [
        (r'\bM[\/\.\s]*S[\/\.\s]*\b', ''),  # Variants of M/s, M. S., etc.
        (r'\bPvt\s*\.?\s*Ltd\.?\b', 'Private Limited'),
        (r'\bPvt[\.]?\b', 'Private'),
        (r'\bLtd[\.]?\b', 'Limited'),
        (r'\bCo[\.]?\b', 'Company'),
        (r'\(P\)', 'Private'),
        (r'\bPrivatelimited\b', 'Private Limited'),
        (r'\s+', ' '),  # Normalize whitespace
    ]

    for pattern, repl in substitutions:
        name = re.sub(pattern, repl, name, flags=re.IGNORECASE)

    # Lowercase text inside parentheses
    name = re.sub(r'\((.*?)\)', lambda m: f'({m.group(1).lower()})', name)

    return name.strip()


def extract_name_until_suffix(name: str) -> str:
    """
    Extracts text up to and including first occurrence of 'ltd', 'limited', or 'llp'
    """
    pattern = r'^(.*?\b(?:ltd|limited|llp)\b)'
    match = re.search(pattern, name, flags=re.IGNORECASE)
    return match.group(1).strip() if match else name


# Example usage:
raw_name = "M/S Example Pvt .Ltd (INDIA) Services LLP"
processed = normalize_company_name(raw_name)
final_name = extract_name_until_suffix(processed)

print("Normalized:", processed)
print("Trimmed to suffix:", final_name)
```

```python
import json
import csv

# Input and output file paths
input_file = r'D:\Nexensus_Projects\HighCourt\High_court_data\Highcourt_company_payload\2025\Bank\Allahabad_High_Court.txt'
output_file = 'output_company_names.csv'

# Use a set to store unique names
unique_company_names = set()

# Read and process each JSON line
with open(input_file, 'r', encoding='utf-8') as file:
    for line in file:
        try:
            data = json.loads(line)
            petitioner = data.get('petitionerName')
            respondent = data.get('respondentName')

            if petitioner:
                unique_company_names.add(petitioner.strip())
            if respondent:
                unique_company_names.add(respondent.strip())
        except json.JSONDecodeError:
            print(f"Skipping invalid JSON line: {line}")

# Write to CSV
with open(output_file, 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['company_name']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for name in sorted(unique_company_names):  # Optional: sorted alphabetically
        writer.writerow({'company_name': name})

print(f"✅ Done! {len(unique_company_names)} unique names saved to {output_file}")
```
